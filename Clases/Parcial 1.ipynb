```python
A=1
B=2
C=0
k=0.15
dt=0.1
t=0
x=[]
yA=[]
yB=[]
yC=[]
while t<100:
    x.append(t)
    yA.append(A)
    yB.append(B)
    yC.append(C)
    dA=-k*A*dt
    dB=+k*A*dt
    dC=k*B*dt
    A+=dA
    B+=dB
    C+=dC
    t+=dt
plt.plot(x,yA)
plt.plot(x,yB)
plt.plot(x,yC)
plt.xlabel("t(min)")
plt.ylabel("Concentración")
plt.title("Concentración vs Tiempo")
plt.show()
```
